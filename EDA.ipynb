```python
#load json data. 
#https://stackoverflow.com/questions/21058935/python-json-loads-shows-valueerror-extra-data

import json    

data = []
with open('Ex-drug.json',encoding="utf8") as f:
    for line in f:
        data.append(json.loads(line))
```

```python
#load json into a dataframe
import pandas as pd
#https://stackoverflow.com/questions/41168558/python-how-to-convert-json-file-to-dataframe

df = pd.DataFrame.from_dict(data, orient='columns')
```

```python
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer

lemtzer = WordNetLemmatizer()

def lemmatize_stemming(text):
    return lemtzer.lemmatize(text, pos='n')

# Write a function to perform the pre processing steps on the entire dataset
def preprocess(text):
    result=[]
    for token in simple_preprocess(text) :
        if token not in STOPWORDS and len(token) > 3:
            result.append(token)
            
    return result

processed_docs  = []

for doc in df["text"]:
    processed_docs.append(preprocess(doc))

#processed_docs

import gensim
dictionary = gensim.corpora.Dictionary(processed_docs)

        

bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
bow_corpus[300]

processed_docs

```

```python
%%time
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=2,id2word=dictionary, passes=30,eta=2)
#lda_model.save('lda.model')
```

```python

for idx, topic in lda_model.print_topics(-2):
    print('Topic: {} Word: {}\n'.format(idx, topic))

# one topic shows people's opinion on this topic 
# one topic shows people;s attitude towards VICE Media
```

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
```

### sentiment analysis

```python
from textblob import TextBlob


def sentiment_func(x):
    sentiment = TextBlob(x['text'])
    x['polarity'] = sentiment.polarity
    x['subjectivity'] = sentiment.subjectivity
    return x

sample = df.apply(sentiment_func, axis=1)

sample.head(10)

```
